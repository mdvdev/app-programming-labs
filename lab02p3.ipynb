```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Fortnite Statistics.csv')

data = data.dropna()

plt.figure(figsize=(10, 6))
sns.barplot(x=data['Placed'], y=data['Eliminations'], errorbar=None)

plt.title('Среднее количество исключений в зависимости от позиции')
plt.xlabel('Позиция в игре (Placed)')
plt.ylabel('Среднее количество исключений (Eliminations)')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data['Damage to Players'], bins=20, kde=False)

plt.xlabel('Урон по игрокам')
plt.ylabel('Количество игроков')
plt.title('Распределение урона по игрокам')

plt.grid(axis='y')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Damage to Players', y='Damage Taken', data=data)

plt.title('Зависимость между нанесенным и полученным уроном')
plt.xlabel('Урон по игрокам (Damage to Players)')
plt.ylabel('Полученный урон (Damage Taken)')
plt.grid(True)
plt.show()
```
