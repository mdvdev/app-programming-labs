```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('IRIS.csv')

print(data.head())

plt.figure(figsize=(10, 8))
# График рассеяния
sns.pairplot(data, hue='species')
plt.title('Парная диаграмма для датасета IRIS')
plt.show()
```

```python
plt.figure(figsize=(12, 12))

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    # Скрипчатая диаграмма
    sns.violinplot(data=data, x='species', y=feature)
    plt.title(f'Скрипичная диаграмма для {feature}')
    plt.xlabel('Вид ириса')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```
