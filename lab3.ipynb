```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AffinityPropagation, MeanShift, AgglomerativeClustering, DBSCAN, estimate_bandwidth
from sklearn.preprocessing import StandardScaler
import time


def load_data(file_path):
    data = np.loadtxt(file_path)
    return data


def plot_clusters(data, labels, title):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=data[:, 0], y=data[:, 1], hue=labels, palette='tab10', legend='full', s=50)
    plt.title(title)
    plt.legend()
    plt.show()


data = load_data('s1.txt')

data = StandardScaler().fit_transform(data)
```

```python
# KMeans
start_time = time.time()
kmeans = KMeans(n_clusters=15)
kmeans_labels = kmeans.fit_predict(data)
end_time = time.time()

print(f"KMeans completed in {end_time - start_time:.2f} seconds")
plot_clusters(data, kmeans_labels, 'KMeans Clustering')
```

```python
# AffinityPropagation
start_time = time.time()
affinity = AffinityPropagation(preference=-50, damping=0.9)
affinity_labels = affinity.fit_predict(data)
end_time = time.time()

print(f"AffinityPropagation completed in {end_time - start_time:.2f} seconds")
plot_clusters(data, affinity_labels, 'AffinityPropagation Clustering')
```

```python
# MeanShift
start_time = time.time()
bandwidth = estimate_bandwidth(data, quantile=0.2) 
meanshift = MeanShift(bandwidth=bandwidth)
meanshift_labels = meanshift.fit_predict(data)
end_time = time.time()

print(f"MeanShift completed in {end_time - start_time:.2f} seconds")
plot_clusters(data, meanshift_labels, 'MeanShift Clustering')
```

```python
# AgglomerativeClustering
start_time = time.time()
agglomerative = AgglomerativeClustering(n_clusters=15)
agglomerative_labels = agglomerative.fit_predict(data)
end_time = time.time()

print(f"AgglomerativeClustering completed in {end_time - start_time:.2f} seconds")
plot_clusters(data, agglomerative_labels, 'Agglomerative Clustering')
```

```python
# DBSCAN
start_time = time.time()
eps = 0.08
min_samples = 5
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(data)
end_time = time.time()

print(f"DBSCAN completed in {end_time - start_time:.2f} seconds")
plot_clusters(data, dbscan_labels, 'DBSCAN Clustering')
```
